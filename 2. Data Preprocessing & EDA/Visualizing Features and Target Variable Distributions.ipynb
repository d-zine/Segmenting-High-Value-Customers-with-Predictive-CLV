```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import numpy as np
```

```python
sbux_clv=pd.read_csv('sbux_clv_with_viewed_completed_channel.csv')
```

```python
sbux_clv.head()
```

```python
#Age
# Extract the column data for 'Age'
data = sbux_clv['Age'][sbux_clv['Age']!="Unknown"]

# Plot distribution plot
sns.histplot(data, bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')

plt.show()
#most users that purchased from Day 1-15 are 51-67, followed by 35-50, then 68-84
```

```python
#Gender
data = sbux_clv['Gender'][(sbux_clv['Gender'] != "Unknown") & (sbux_clv['Gender'] != "O")]


# Plot distribution plot
sns.histplot(data, bins=10)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Gender')

plt.show()
#more males than females purchased from Day 1-15
```

```python
### Define the desired order of the income categories
income_order = ['30k-50k', '50k-75k', '75k-100k', '100k-120k']

# Extract the column data for 'HH Income' excluding "Unknown"
data = sbux_clv['HH Income'][(sbux_clv['HH Income'] != "Unknown")]

# Convert the data to a categorical type with the specified order
data = pd.Categorical(data, categories=income_order, ordered=True)

# Plot distribution plot
sns.histplot(data)
plt.xlabel('HH Income')
plt.ylabel('Frequency')
plt.title('Distribution of HH Income')

plt.show()
#most users who've purchased from Day 1-15 make 50k-75k
```

```python
# Extract the column data for 'MemberSince'
data = sbux_clv['MemberSince']

# Plot bar plot
sns.countplot(x=data)
plt.xlabel('MemberSince')
plt.ylabel('Frequency')
plt.title('Distribution of MemberSince')

plt.show()
#Most members who've purchased from Day 1-15 signed up in 2017
```

```python
#Recency (# Days ago from last trans. As of final day up to Day 15)

# Extract the column data for 'MemberSince'
data = sbux_clv['Recency (# Days ago from last trans. As of final day up to Day 15)']

# Plot bar plot
sns.countplot(x=data)
plt.xlabel('Recency (# Days ago from last trans. As of final day up to Day 15)')
plt.ylabel('Frequency')
plt.title('Distribution of Recency (# Days ago from last trans. \nAs of final day up to Day 15)')

plt.show()
#Most people have purchased as recent as the final day of measurement (Day 15)
```

```python
#Frequency (# Trans. from Day 1-15)

data = sbux_clv['Frequency (# Trans. from Day 1-15)']

# Plot bar plot
sns.countplot(x=data)
plt.xlabel('Frequency (# Trans. from Day 1-15)')
plt.ylabel('Frequency')
plt.title('Distribution of Frequency (# Trans. from Day 1-15)')

plt.show()
#Some people have purchased more than once per day from Day 1-15, inferred since the frequency ranges from 1-20
```

```python
data = sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)']

# Plot density plot
plt.figure(figsize=(16, 6))
sns.kdeplot(data)
plt.xlabel('Monetary Value (Sum of Trans. from Day 1-15)')
plt.ylabel('Density')
plt.title('Distribution of Monetary Value (Sum of Trans. from Day 1-15)')

plt.show()
#right skewed

# Compute descriptive statistics
statistics = data.describe()
print(f"Statistics for Monetary Value (Sum of Trans. from Day 1-15):\n{statistics}\n")

# Compute descriptive statistics
statistics = data.describe()
statistics['median'] = data.median()  # Adding median to the statistics
print(f"Statistics for Monetary Value (Sum of Trans. from Day 1-15):\n{statistics}\n")
statistics 
```

```python
# Log transform the data
log_data = np.log1p(data)

# Plot density plot
plt.figure(figsize=(16, 6))
sns.kdeplot(log_data)
plt.xlabel('Monetary Value (Sum of Trans. from Day 1-15)')
plt.ylabel('Density')
plt.title('Distribution of Log-Transformed Monetary Value (Sum of Trans. from Day 1-15)')

plt.show()
```
