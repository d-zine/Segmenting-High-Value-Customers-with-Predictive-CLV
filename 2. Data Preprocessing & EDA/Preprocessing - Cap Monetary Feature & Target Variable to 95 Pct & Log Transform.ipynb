###### Anything >= 95% for 'Monetary Value (Sum of Trans. from Day 1-15)' and 'Sum(Trans. Amt from Day 16-30)', impute with 95% number

```python
# Calculate the 95th percentile for the columns
monetary_value_95th = sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'].quantile(0.95)
trans_amt_95th = sbux_clv['Sum(Trans. Amt from Day 16-30)'].quantile(0.95)

# Impute values equal or above the 95th percentile with the 95th percentile value
sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'] = sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'].where(
    sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'] < monetary_value_95th, monetary_value_95th)
sbux_clv['Sum(Trans. Amt from Day 16-30)'] = sbux_clv['Sum(Trans. Amt from Day 16-30)'].where(
    sbux_clv['Sum(Trans. Amt from Day 16-30)'] < trans_amt_95th, trans_amt_95th)
```

```python
# Calculate the 95th and 99th percentiles using describe()
percentiles = sbux_clv.describe(percentiles=[0.95, 0.99])

# Print the percentiles
percentiles
```

```python
# Fill NaNs, zeros, and negative values with a small positive value
sbux_clv['Sum(Trans. Amt from Day 16-30)'].fillna(1e-8, inplace=True)
sbux_clv['Sum(Trans. Amt from Day 16-30)'].replace(0, 1e-8, inplace=True)
sbux_clv['Sum(Trans. Amt from Day 16-30)'].clip(lower=1e-8, inplace=True)

sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'].fillna(1e-8, inplace=True)
sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'].replace(0, 1e-8, inplace=True)
sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'].clip(lower=1e-8, inplace=True)

# Log-transform 'Sum(Trans. Amt from Day 16-30)' column
sbux_clv['Sum(Trans. Amt from Day 16-30)'] = np.log(sbux_clv['Sum(Trans. Amt from Day 16-30)'])

# Log-transform 'Monetary Value (Sum of Trans. from Day 1-15)' column
sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'] = np.log(sbux_clv['Monetary Value (Sum of Trans. from Day 1-15)'])
sbux_clv.describe()
```
